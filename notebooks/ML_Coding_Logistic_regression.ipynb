# Config and Parameters

- lr: learning rate
- max_iter: maximum iterations
- tol: tolerance for stop using gradient norm

# Initialization
- Init weights and bias to zero

# supporting functions
- sigmoid: 1/(1+exp(-z))

# Fitting
- forward prop: y_hat = sigmoid(w*X+b)
- Loss: cross entropy: L = -1/n sum(y_i log(y_hat_i) + (1-y_i)*log(1-y_hat_i))
- Gradient: dw = (1/n) X^T*err , db = mean(error)
- Gradient descentï¼š w = w - lr*dw, b = b-lr*db
- stop: iter> max_iter or |dw|<tol

# Prediction
- yhat = sigmoid(X*w+b)

```python
from ml_algorithm_playground.linear.logistic_regression import LogisticRegression
```

```python
n_samples,n_features = 10,4
seed= 42

import numpy as np

# Set a random seed for reproducibility
np.random.seed(42) # You can use any integer here

# Create a (5,4) numpy array with random values between 0 and 1
X = np.random.rand(n_samples, n_features)
print("Deterministic Random Array (using seed 42):\n", X)
```

```python
Y = np.random.randint(0,2,n_samples)
print("Deterministic Random Array (using seed 42):\n", Y)
```

```python
model=LogisticRegression(treshold=0.5,lr=0.01, max_iter=1000, tol=1e-4)
```

```python
model.fit(X,Y)
```

```python
y_hat= model.predict(X)
print(f"y_hat: {y_hat}")
```
