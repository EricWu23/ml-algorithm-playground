b: (q,) : np.zeros(q)

# fit(X,y)
## forward
z= X*W+b
y_hat= identity(z)
## gradient
err= y_hat-y
dw = 1/n*(err)X
db = np.mean(err)
## update
W= W-dw
b = b-db
