# Test

```python
N =20
D = 4
```

```python
import numpy as np

# Set a random seed for reproducibility
np.random.seed(42) # You can use any integer here

# Create a (5,4) numpy array with random values between 0 and 1
X = np.random.rand(N, D)
print("Deterministic Random Array (using seed 42):\n", X)
```

```python
clusters = Kmeans(k=4, max_iters=100, tol=0, rng = np.random.default_rng(seed=42))
```

```python
clusters.fit(X)
```

```python
clusters.tol
```

```python
clusters._shift
```

```python
clusters._iteration_counter
```

```python
Y=clusters.predict(X)
```

```python
Y
```
